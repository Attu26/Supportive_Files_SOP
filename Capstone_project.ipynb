**Created a user definded function for finding the lower and upper range for a variable so that outlier can be treated.**

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lrincome,urincome=remove_outlier(df1['Yearly_avg_view_on_travel_page'])
df1['Yearly_avg_view_on_travel_page']=np.where(df1['Yearly_avg_view_on_travel_page']>urincome,urincome,df1['Yearly_avg_view_on_travel_page'])
df1['Yearly_avg_view_on_travel_page']=np.where(df1['Yearly_avg_view_on_travel_page']<lrincome,lrincome,df1['Yearly_avg_view_on_travel_page'])


lrtravel,urtravel=remove_outlier(df1['total_likes_on_outstation_checkin_given'])
df1['total_likes_on_outstation_checkin_given']=np.where(df1['total_likes_on_outstation_checkin_given']>urtravel,urtravel,df1['total_likes_on_outstation_checkin_given'])
df1['total_likes_on_outstation_checkin_given']=np.where(df1['total_likes_on_outstation_checkin_given']<lrtravel,lrtravel,df1['total_likes_on_outstation_checkin_given'])

lrmiles,urmiles=remove_outlier(df1['Yearly_avg_comment_on_travel_page'])
df1['Yearly_avg_comment_on_travel_page']=np.where(df1['Yearly_avg_comment_on_travel_page']>urmiles,urmiles,df1['Yearly_avg_comment_on_travel_page'])
df1['Yearly_avg_comment_on_travel_page']=np.where(df1['Yearly_avg_comment_on_travel_page']<lrmiles,lrmiles,df1['Yearly_avg_comment_on_travel_page'])

lrmiles,urmiles=remove_outlier(df1['member_in_family'])
df1['member_in_family']=np.where(df1['member_in_family']>urmiles,urmiles,df1['member_in_family'])
df1['member_in_family']=np.where(df1['member_in_family']<lrmiles,lrmiles,df1['member_in_family'])

lrmiles,urmiles=remove_outlier(df1['total_likes_on_outofstation_checkin_received'])
df1['total_likes_on_outofstation_checkin_received']=np.where(df1['total_likes_on_outofstation_checkin_received']>urmiles,urmiles,df1['total_likes_on_outofstation_checkin_received'])
df1['total_likes_on_outofstation_checkin_received']=np.where(df1['total_likes_on_outofstation_checkin_received']<lrmiles,lrmiles,df1['total_likes_on_outofstation_checkin_received'])

lrmiles,urmiles=remove_outlier(df1['week_since_last_outstation_checkin'])
df1['week_since_last_outstation_checkin']=np.where(df1['week_since_last_outstation_checkin']>urmiles,urmiles,df1['week_since_last_outstation_checkin'])
df1['week_since_last_outstation_checkin']=np.where(df1['week_since_last_outstation_checkin']<lrmiles,lrmiles,df1['week_since_last_outstation_checkin'])

lrmiles,urmiles=remove_outlier(df1['montly_avg_comment_on_company_page'])
df1['montly_avg_comment_on_company_page']=np.where(df1['montly_avg_comment_on_company_page']>urmiles,urmiles,df1['montly_avg_comment_on_company_page'])
df1['montly_avg_comment_on_company_page']=np.where(df1['montly_avg_comment_on_company_page']<lrmiles,lrmiles,df1['montly_avg_comment_on_company_page'])



lrmiles,urmiles=remove_outlier(df1['Daily_Avg_mins_spend_on_traveling_page'])
df1['Daily_Avg_mins_spend_on_traveling_page']=np.where(df1['Daily_Avg_mins_spend_on_traveling_page']>urmiles,urmiles,df1['Daily_Avg_mins_spend_on_traveling_page'])
df1['Daily_Avg_mins_spend_on_traveling_page']=np.where(df1['Daily_Avg_mins_spend_on_traveling_page']<lrmiles,lrmiles,df1['Daily_Avg_mins_spend_on_traveling_page'])

df.shape
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
  vif = pd.DataFrame()
  vif["variables"] = X.columns
  vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

  return(vif)
```

```python

calc_vif(df1).sort_values(by = 'VIF', ascending = False)
```

Though most of the variables are having their VIF less than 10, there are some variables like Yearly average view on travel page, monthly average comment oncompany page and preferred location type label which have inflation greater than 10.

So, we will drop these variables one by one till we have all the variables whose variance inflation factor less than 10.

follwoing are the variable variables after we drop some.

```python
df2 = df1.drop(['preferred_device','preferred_location_type','following_company_page','working_flag'],axis=1)
```

```python
def calculate_vif(X, thresh = 10):
  variables = list(range(X.shape[1]))
  dropped = True
  while dropped:
    dropped = False
    vif = [variance_inflation_factor(X.iloc[:, variables].values, ix)
    for ix in range(X.iloc[:, variables].shape[1])]

    maxloc = vif.index(max(vif))
    if max(vif) >= thresh:
      print('dropping \'' + X.iloc[:, variables].columns[maxloc])
      del variables[maxloc]
      dropped = True

      print('Remaining variables:')
      print(X.columns[variables])
      return X.iloc[:, variables]
```
